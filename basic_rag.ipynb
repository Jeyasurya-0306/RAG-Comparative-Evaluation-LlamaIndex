```python
import requests

url = "https://licindia.in/documents/20121/62876/Final+Policy+document_LICs+New+Jeevan+Shanti_V05_logo.pdf/f7950a05-1136-1bfb-bf4a-f997d85fc30f?t=1707288544327"
pdf_path = "LIC_New_Jeevan_Shanti.pdf"

response = requests.get(url)
with open(pdf_path, "wb") as f:
    f.write(response.content)

print("PDF downloaded successfully:", pdf_path)
```

```python
from llama_index.readers.file import PDFReader

# Load PDF
reader = PDFReader()
documents = reader.load_data(pdf_path)
print("Parsed", len(documents), "documents")

# Example: check first document
print(documents[0].text)
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

embedding_models = {
    "MiniLM-MSMARCO": HuggingFaceEmbedding(model_name="sentence-transformers/msmarco-MiniLM-L6-cos-v5"),
    "DistilBERT-MSMARCO": HuggingFaceEmbedding(model_name="sentence-transformers/msmarco-distilbert-base-v4"),
    "Sentence-T5-Large": HuggingFaceEmbedding(model_name="sentence-transformers/sentence-t5-large")
}
```

```python
import faiss
import numpy as np
from llama_index.vector_stores.faiss import FaissVectorStore
from llama_index.core import VectorStoreIndex, StorageContext

vector_indexes = {}

for name, embed_model in embedding_models.items():
    print(f"Building FAISS index for {name}...")

    dim = np.array(embed_model.get_text_embedding("hello")).shape[0]
    faiss_index = faiss.IndexFlatL2(dim)
    vector_store = FaissVectorStore(faiss_index=faiss_index)
    storage_context = StorageContext.from_defaults(vector_store=vector_store)

    index = VectorStoreIndex.from_documents(
        documents,
        storage_context=storage_context,
        embed_model=embed_model
    )
    vector_indexes[name] = index
print("all indexes are build done ")
```

```python
query = "what are the definitions of terms/words used in the Policy?"
```

```python
index_to_use = vector_indexes[  "DistilBERT-MSMARCO"]
retriever = index_to_use.as_retriever(similarity_top_k=1)
retrieved_nodes = retriever.retrieve(query)
context_text = "\n\n".join([n.get_text() for n in retrieved_nodes])
```

```python
from transformers import pipeline

# Use Flan-T5-base
flan = pipeline("text2text-generation", model="google/flan-t5-base", device=-1)

prompt = f"""You are a strict teacher. Read the context carefully and locate the exact part
that answers the question. Copy the relevant sentence(s) word-for-word from the context.

Context:
{context_text}

Question: {query}
Answer:
"""

flan_answer = flan(prompt,
                   max_new_tokens=512,
                   truncation=True
                   )[0]['generated_text']
print("Flan-T5-base Answer:\n", flan_answer)
```

```python
opt_model = pipeline("text-generation", model="facebook/opt-350m", device=-1)

# the prompt
prompt_opt = f"""You are a strict teacher. Read the context carefully and locate the exact part
that answers the question. Copy the relevant sentence(s) word-for-word from the context.

Context:
{context_text_truncated}

Question: {query}
Answer:"""

answer_opt = opt_model(
    prompt_opt,
    max_new_tokens=320,
    do_sample=False,
    return_full_text=False
)[0]["generated_text"]

print("OPT-350M Answer:\n", answer_opt)
```
